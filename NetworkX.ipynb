```python
import matplotlib.pyplot as plt
import networkx as nx

G = nx.DiGraph()

G.add_edge('BTC-USD','ETH-USD')
G.add_edge('ETH-USD','BTC-USD')
G.add_edge('3','2')
G.add_edge('3','4')
G.add_edge('4','3')
G.add_edge('3','1')

nx.draw(G, node_color='y', edgecolors='k', width=2.0, with_labels=True)
plt.show()
```

```python
import networkx as nx
G=nx.complete_graph(8)
G.edges()
nodenames = ['BTC-USD', 'ETH-USD', 'LTC-USD', 'USDT-USD', 'USDC-USD', 'BNB-USD','XRP-USD', 'XMR-USD']
mapping = {i:nodename for i,nodename in enumerate(nodenames)}
#now mapping looks like {0:'a', 1:'b', ...}
H=nx.relabel_nodes(G,mapping)
H.edges()
nx.draw(H, node_color='b', edgecolors='k', width=2.0, with_labels=True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import networkx as nx
import networkx as nx

g1 = nx.Graph()


G = nx.DiGraph()
array = np.random.randint(1, 100, size=(8, 8))
print(array)
```

```python

columns = ['BTC-USD', 'ETH-USD', 'LTC-USD', 'USDT-USD', 'USDC-USD', 'BNB-USD','XRP-USD', 'XMR-USD']
df = pd.DataFrame(array,columns=columns,index = columns)


print(df)
print(type(df))
```

```python


g2 = nx.Graph()
for column in columns:
    for index, row in df.iterrows():
        print(column,row[column],index)
        g2.add_edge(column, index, weight = row[column])
```

```python

nx.draw(g2, node_color='b',  with_labels=True)
plt.show()
```

```python
cat1 = [(u, v) for (u, v, d) in g2.edges(data=True) if d['weight'] <= 10]
cat2 = [(u, v) for (u, v, d) in g2.edges(data=True) if d['weight'] <= 20 and d['weight'] >= 11]
cat3 = [(u, v) for (u, v, d) in g2.edges(data=True) if d['weight'] <= 30 and d['weight'] >= 21]
cat4 = [(u, v) for (u, v, d) in g2.edges(data=True) if d['weight'] <= 50 and d['weight'] >= 31]
cat5 = [(u, v) for (u, v, d) in g2.edges(data=True) if d['weight'] <= 70 and d['weight'] >= 51]
cat6 = [(u, v) for (u, v, d) in g2.edges(data=True) if d['weight'] > 70]
```

```python
pos = nx.spring_layout(g2)  # positions for all nodes
```

```python
nx.draw_networkx_nodes(g2, pos, node_size=1000)
```

```python
nx.draw_networkx_nodes(g2, pos,label=columns,node_size=500)
nx.draw_networkx_edges(g2, pos,label=columns, edgelist=cat1, edge_color='b')
nx.draw_networkx_edges(g2, pos,label=columns, edgelist=cat2, edge_color='r')
nx.draw_networkx_edges(g2, pos,label=columns, edgelist=cat3, edge_color='g')
nx.draw_networkx_edges(g2, pos,label=columns, edgelist=cat4, edge_color='k')
nx.draw_networkx_edges(g2, pos,label=columns, edgelist=cat5, edge_color='#1f78b4')
nx.draw_networkx_edges(g2, pos,label=columns, edgelist=cat6, edge_color='#1f08b4')


nx.draw_networkx_labels(g2,pos,font_size=16,font_color='k')
```
